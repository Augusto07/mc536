## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import io
import requests
import json

url_miracle = "https://dbpedia.org/data/Tent_of_Miracles_(novel).jsod"
data_miracle = requests.get(url_miracle)
json_miracle = data_miracle.json()
print(json.dumps(json_miracle, indent=2))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
author = json_miracle["d"]["results"][0]["http://dbpedia.org/ontology/author"]["__deferred"]["uri"]
print("O autor de Tenda dos Milagres é: "+author)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
#Getting URL
concept = "Aspirin"
match = "exact"
limit = "10"

url_aspirin = f"https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"
concept = concept.replace(" ", "%20")
data = requests.get(url_aspirin)
json_mi = data.json()

#Getting code
url_code = json_mi[0]["resource"]
code = url_code[str.rindex(url_code, "/")+1:]

#Accessing Code URL
url_template = f"https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"
data = requests.get(url_template)
json_mid = data.json()

#Printing terms
for x in json_mid["terms"]:
    print(x["label"])
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
format = "json"
query = "Sydney Opera House"

url_opera = f"http://nominatim.openstreetmap.org/search?format={format}&q={query}"

data = requests.get(url_opera)
opera_ref = data.json()
print(json.dumps(opera_ref, indent=2))
```

```python
#printing relation
id = str(opera_ref[0]["osm_id"])
element_type = "relation"
format = "json"
url_opera = f"https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

print(url_opera)
opera_data = requests.get(url_opera)
opera_relation = opera_data.json()
opera_dumps = json.dumps(opera_relation, indent=2)
print(opera_dumps)
```

```python
dbpedia_name = opera_relation["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                                 .replace("en:", "")
format = "jsod"

url_dbpedia = f"http://dbpedia.org/data/{dbpedia_name}.{format}"

data_opera = requests.get(url_dbpedia)
opera_dbpedia = data_opera.json(strict=False)
print(json.dumps(opera_dbpedia, indent=2))
```
